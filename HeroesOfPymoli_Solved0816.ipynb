```python
import pandas as pd
import numpy as np
```

```python
file='C:\workspace\Resources\purchase_data.csv'
```

```python
df=pd.read_csv(file)
```

```python
### Player Count
df_a=len(df['SN'].unique())
df_aa=pd.DataFrame({'Total Player':[df_a]})
df_aa
```

```python
### Purchasing Analysis (total)
dfa=len(df['Item Name'].unique())
dfb=round(df['Price'].mean(),2)
dfc=len(df['Item ID'])
dfd=df['Price'].sum()
df_pur_ana=pd.DataFrame({'Number of Unique Items':[dfa],'Average Price':[dfb],
                        'Number of Purchases':[dfc], 'Total Revenue':[dfd]})
df_pur_ana['Average Price']=df_pur_ana['Average Price'].map('${:.2f}'.format)
df_pur_ana['Total Revenue']=df_pur_ana['Total Revenue'].map('${:,.2f}'.format)
df_pur_ana
```

```python
### Gender Demographics
df_unique_player=df.drop_duplicates('SN',keep='first',inplace=False)
df_males=len(df_unique_player.loc[df_unique_player['Gender']=='Male',:])
df_females=len(df_unique_player.loc[df_unique_player['Gender']=='Female',:])
df_other=len(df_unique_player.loc[df_unique_player['Gender']=='Other / Non-Disclosed',:])
df_males_pct=round(df_males/(df_males+df_females+df_other)*100,2)
df_females_pct=round(df_females/(df_males+df_females+df_other)*100,2)
df_other_pct=round(df_other/(df_males+df_females+df_other)*100,2)
df_gender=pd.DataFrame({'Total Count':[df_males,df_females,df_other],
                        'Percentage of Players':[df_males_pct,df_females_pct,df_other_pct]},
                        index=['Male','Female','Other / Non-Disclosed'])
df_gender
```

```python
### Purchasing Analysis (Gender)
df_pur_count=df.groupby(df['Gender'])['Purchase ID'].count()
df_ave_price=df.groupby(df['Gender'])['Price'].mean()
df_total_value=df.groupby(df['Gender'])['Price'].sum()
df_ave_pur=round(df_total_value/df_gender['Total Count'],2)
df_pur_ana=pd.DataFrame({'Purchase Count':df_pur_count,
                         'Average Purchase Price':df_ave_price,
                         'Total Purchase Value':df_total_value},
                        index=['Male','Female','Other / Non-Disclosed'])
df_pur_ana['Avg Total Purchase per Person']=df_ave_pur
df_pur_ana['Average Purchase Price']=df_pur_ana['Average Purchase Price'].map('${:.2f}'.format)
df_pur_ana['Total Purchase Value']=df_pur_ana['Total Purchase Value'].map('${:.2f}'.format)
df_pur_ana['Avg Total Purchase per Person']=df_pur_ana['Avg Total Purchase per Person'].map('${:.2f}'.format)
df_pur_ana
```

```python
### Age Demographics
bins=[0,9,14,19,24,29,34,39,100]
group_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
df_unique_player['Age Group']=pd.cut(df_unique_player['Age'],bins,labels=group_names)
df_total_count=df_unique_player.groupby('Age Group')['Purchase ID'].count().reset_index()
df_age_demo=df_total_count.rename(columns={'Purchase ID':'Total Count'})
pct_player=round(df_age_demo['Total Count']/df_a*100,2)
df_age_demo['Percentage of Players']=pct_player
df_age_demo_reset=df_age_demo.set_index("Age Group")
df_age_demo_reset
```

```python
### Purchasing Analysis (Age)
df['Age Group']=pd.cut(df['Age'],bins,labels=group_names)
df_total_count2=df.groupby('Age Group')['Purchase ID'].count().reset_index()
df_age_purchase=df_total_count2.rename(columns={'Purchase ID':'Purchase Count'})
aver_purc_pric=round(df.groupby('Age Group')['Price'].mean().reset_index(),2)
tota_purc_valu=round(df.groupby('Age Group')['Price'].sum().reset_index(),2)
df_age_purchase_fin=pd.merge(df_age_purchase,aver_purc_pric,
                             on="Age Group")
df_age_purchase_fina=pd.merge(df_age_purchase_fin,tota_purc_valu,
                              on="Age Group")
df_age_purchase_final=df_age_purchase_fina.rename(columns={'Price_x':'Average Purchase Price',
                                                          'Price_y':'Total Purchase Value'})

df_age_purchase_final_reset=df_age_purchase_final.set_index("Age Group")
df_age_purchase_final_reset['Total Purchase Value']
ave_totals=round(df_age_purchase_final_reset['Total Purchase Value']/df_age_demo_reset['Total Count'],2)
df_age_purchase_final_reset['Avg Total Purchase per Person']=ave_totals
df_age_purchase_final_reset['Average Purchase Price']=df_age_purchase_final_reset['Average Purchase Price'].map('${:.2f}'.format)
df_age_purchase_final_reset['Total Purchase Value']=df_age_purchase_final_reset['Total Purchase Value'].map('${:.2f}'.format)
df_age_purchase_final_reset['Avg Total Purchase per Person']=df_age_purchase_final_reset['Avg Total Purchase per Person'].map('${:.2f}'.format)
df_age_purchase_final_reset
```

```python
### Top Spenders
df['Total Purchase Value'] = df.groupby(['SN'])['Price'].transform('sum')
df['Purchase Count']=df.groupby(['SN'])['Purchase ID'].transform('count')
df['Average Purchase Price']=round(df['Total Purchase Value']/df['Purchase Count'],2)
df_top_spender = df.drop_duplicates(subset=['SN'])

df_top_spender_sort=df_top_spender.sort_values('Total Purchase Value',ascending=False).head()
df_top_spender_fin=df_top_spender_sort[['SN','Purchase Count',
                                          'Average Purchase Price',
                                          'Total Purchase Value',]]
df_top_spender_finale=df_top_spender_fin.set_index('SN')
df_top_spender_finale['Average Purchase Price']=df_top_spender_finale['Average Purchase Price'].map('${:.2f}'.format)
df_top_spender_finale['Total Purchase Value']=df_top_spender_finale['Total Purchase Value'].map('${:.2f}'.format)
df_top_spender_finale
```

```python
### Most Popular Items
df['Purchase Count Items'] = df.groupby(['Item Name'])['Item ID'].transform('count')
df['Total Value Items']=round(df['Purchase Count Items']*df['Price'],2)
df_top_items = df.drop_duplicates(subset=['Item Name'])
df_top_items_sort=df_top_items.sort_values('Purchase Count Items',ascending=False).head()
df_top_items_fin=df_top_items_sort[['Item ID','Item Name','Purchase Count Items',
                                    'Price','Total Value Items']]
df_top_items_fina=df_top_items_fin.set_index(['Item ID','Item Name'])
df_top_items_finale=df_top_items_fina.rename(columns={'Purchase Count Items':'Purchase Count',
                                                     'Price':'Item Price',
                                                     'Total Value Items':'Total Purchase Value'})
df_top_items_finale['Item Price']=df_top_items_finale['Item Price'].map('${:.2f}'.format)
df_top_items_finale['Total Purchase Value']=df_top_items_finale['Total Purchase Value'].map('${:.2f}'.format)
df_top_items_finale
```

```python
### Most Profitable Items
df_top_value_sort=df_top_items.sort_values('Total Value Items',ascending=False).head()
df_top_value_fin=df_top_value_sort[['Item ID','Item Name','Purchase Count Items',
                                    'Price','Total Value Items']]
df_top_value_fina=df_top_value_fin.set_index(['Item ID','Item Name'])
df_top_value_finale=df_top_value_fina.rename(columns={'Purchase Count Items':'Purchase Count',
                                                     'Price':'Item Price',
                                                     'Total Value Items':'Total Purchase Value'})
df_top_value_finale['Item Price']=df_top_value_finale['Item Price'].map('${:.2f}'.format)
df_top_value_finale['Total Purchase Value']=df_top_value_finale['Total Purchase Value'].map('${:.2f}'.format)
df_top_value_finale
```
